```python
tracks_df = data_all_[["TRACK_ID", "POSITION_X", "POSITION_Y", "POSITION_T", "FRAME"]]

fig, ax = plt.subplots(figsize=(15,15))

#print(tracks_df["track_id"].unique())
for track_id in tracks_df["TRACK_ID"].unique():
    single_track_df = tracks_df[tracks_df["TRACK_ID"]==track_id]
    single_track_df = single_track_df.sort_values(by="FRAME")
    ax.plot(single_track_df['POSITION_X'], single_track_df['POSITION_Y'])
```

```python
migration_speed_df = pd.DataFrame(columns=tracks_df.columns)

intervals = [1,5,10]

migration_speed_dfs = [migration_speed_df.copy(), migration_speed_df.copy(), migration_speed_df.copy()]

for index, interval in enumerate(intervals):
    for track_id in tracks_df["TRACK_ID"].unique():

        single_track_df = tracks_df[tracks_df ["TRACK_ID"]==track_id]
        single_track_df = single_track_df.sort_values(by="FRAME")
        dist = single_track_df.diff(interval).fillna(0.)

        single_track_df["step_size"] = np.sqrt(dist.POSITION_X**2 + dist.POSITION_Y**2)
        single_track_df["step_size_x"] = dist.POSITION_X
        single_track_df["step_size_y"] =  dist.POSITION_Y
        single_track_df["vel"] = np.sqrt(dist.POSITION_X**2 + dist.POSITION_Y**2)/dist.FRAME
        single_track_df["vel_x"] = dist.POSITION_X/dist.FRAME
        single_track_df["vel_y"] =  dist.POSITION_Y/dist.FRAME
        single_track_df["phi"] =  np.arctan2(dist.POSITION_Y,-dist.POSITION_X)*180.0/np.pi
        
        #print("Track ID %s" % track_id)
        #print(single_track_df.head())
        migration_speed_dfs[index] = migration_speed_dfs[index].append(single_track_df)
```

```python
plot_migration_speeds = []
for i in range(len(intervals)):
    plot_migration_speeds.append(migration_speed_dfs[i].dropna())
    plot_migration_speeds[i] = plot_migration_speeds[i][plot_migration_speeds[i]["FRAME"] > intervals[i]]
```

```python
fig, ax = plt.subplots(len(intervals),figsize=(15,30))
for i in range(len(intervals)):
    sns.lineplot(x = "POSITION_T", y = "vel_x", data = plot_migration_speeds[i], ax=ax[i]) 
    interval_length_min = 5*intervals[i]
    ax[i].set_title("time interval %s min" % interval_length_min)
    ax[i].set_ylabel("velocity in microns/min")
```
